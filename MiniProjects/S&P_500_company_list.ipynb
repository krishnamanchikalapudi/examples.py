# Read S&P 500 company information from wikipedia

```
import pandas as pd

html = pd.read_html("https://en.wikipedia.org/wiki/List_of_S%26P_500_companies")
sp500Companies = html[0] # list 1st table data
sp500Companies
```

## Sort by Symbol

```
sp500Companies.sort_values("Symbol")
```

## Sort by Central Index Key (CIK)

```
sp500Companies.sort_values("CIK")
```

## Sort by name

```
sp500Companies.sort_values("Security")
```

## Sort by founded

```
sp500Companies.sort_values("Founded")
```
